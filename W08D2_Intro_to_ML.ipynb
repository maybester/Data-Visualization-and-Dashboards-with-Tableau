# Types of Machine Learning: Supervised and Unsupervised

## Machine Learning: Supervised Learning
- In supervised learning, we have a set of observations (__*X*__) with an associated target (__*y*__)
- We wish to find a model function that relates __*X*__ to __*y*__
- Then use that model function to predict future observations

# Types of Supervised Learning: Regression and Classification

## Classification vs. Regression

* Classification problems: predicting among two or more categories, also known as classes
    - Predict whether a patient has a liver disease or not
    - Predict whether the letter grade of a student (A, B, C, D or F)
* Regression problem: predicting a continuous (in other words, a number) value
    - Predict house prices
    - Predict someone's age from their photo

## The golden rule
- When you're doing supervised learning, now that you've identified **X** and **y**
- **You need to split your data into train and test**
- **You only work with the training data**

### Why?
- We want to "hide" some data while training, so that we can later evaluate our trained model on **unseen data**.
    - Analogy: studying for a test.
- The ability to evaluate on unseen data is very important. It allows to make decisions like which model is best and/or which features should be included/dropped.
- Compare this to the statistical modelling earlier, where we didn't split our data.
    - There, we were more interested in finding the relationships in our data.
    - For ML, we are more interested in making predictions on future data.

## The big picture
- We train using the **training data**
- We test what is learned by the model on the **test data**
- We have two scores: **training** vs. **test**
- Example:
    - Model predicts apple or banana with a 95% training accuracy (accuracy on the training data)
    - Model predicts apple or banana with a 90% testing accuracy (accuracy on the testing data)

### Which matters more?
- It doesn't matter how good our **training score** is because the **test score is what matters**
- Good models that generalize well though will have **similar training and testing scores**
- Typically training scores are higher than testing scores.
- The difference between the training and testing scores is called **generalization error**.

**We want to pick models that generalize well to unseen data**

## The fundamental tradeoff

- Different ML models can different amounts of "complexity".

| Model | Training Score relative to Test Score | Performance |
|:-|:-|:-|
|Too Simple|Low training score and low test score|Underfit|
| Too Complex|High training score compared to test score| Overfit |


- Models that have **extremely high training scores** (that are too good to be true) that are **highly complex** that learned very complex relationships in the training data **can be overfit**
    - Modeled **noise** in the training data instead of a relationship/pattern.
- On the other hand, models that have **low training scores** that are **very simple** may not have learned the necessary relationships in the training data needed to predict well on unseen data; they are **underfit**

## The fundamental tradeoff (part 2)
In the "bias-variance" language

- **Bias**
    - Bias refers to the error introduced by approximating a real-world problem with a simplified model. A model with high bias error tends to make simplistic assumptions about the underlying data, leading to underfitting.
    - An underfit model is not able to capture the true patterns and relationships in the data, resulting in poor performance.
- **Variance**
    - Measures the variability of a model for different training datasets. A model with high variance is sensitive to fluctuations in the training data and tends to overfit.
    - An overfit model has memorized the training data to a great extent, but it fails to generalize well to unseen data, leading to poor performance on test or validation datasets.
    
- Goal is to find a model that balances bias and variance, so the total error is minimized.

## Errors in Prediction
- **Approximation error**
    - The difference between the true function that maps input data to output values and the approximated function that a learning algorithm or model produces.
- **Estimation error**
    - The difference between the true values of parameters and their estimated values obtained from learning algorithm.
- **Optimization error**
    - Error that occurs during the process of optimization. More on this later when you learn about loss functions and gradient descent.


## Linear Regression

- Linear regression is one of the most basic and popular ML/statistical techniques.
- Used as a predictive model
- Assumes a linear relationship between the dependent variable (which is the variable we are trying to predict/estimate, **y**) and the independent variable/s (input variable/s used in the prediction, **X**)

### Let's start with **simple** linear regression
- Only one independent/input variable is used to predict the dependent variable.

## Simple Linear Regression

$$\hat{y} = wx + b$$

$\hat{y}$ = Dependent variable

$b$ = Constant

$w$ = Coefficients

$x$ = Independent variable

## Multiple Linear Regression
- Many $x$'s and $w$'s

$$\hat{y} = w_1x_1 + w_2x_2 + ... + b$$

- The larger the relative value of $w_i$, the more influence $x_i$ has on the target $\hat{y}$

## Matrix representation

- $\hat{y}$ is the linear function of features $x$ and weights $w$.

$$\hat{y} = w^Tx + b$$
        
- $\hat{y} \rightarrow$ prediction
- $w \rightarrow$ weight vector
- $b \rightarrow$ bias
- $x \rightarrow$ features

$$\hat{y} = \begin{bmatrix}w_1 & w_2 & \cdots & w_d\end{bmatrix}\begin{bmatrix}x_1 \\ x_2 \\ \vdots \\ x_d\end{bmatrix} + b$$

## Matrix representation for multiple predictions

$$\hat{y} = w^TX + b$$
        

$$\hat{y} = \begin{bmatrix}w_1 & w_2 & \cdots & w_d\end{bmatrix}\begin{bmatrix}x^{(1)}_1 & x^{(2)}_1 & \ldots &x^{(n)}_1\\ x^{(1)}_2 & x^{(2)}_2 & \ldots & x^{(n)}_2 \\ \vdots & \vdots & \ldots \\ x^{(1)}_d & x^{(2)}_d & \ldots & x^{(n)}_d\end{bmatrix} + b$$

## Polynomial regression

### Non-linear regression motivation
- Linear regression might seem rather limited.
- What if the true relationship between the target and the features is non-linear?
- A more complex model than linear regression

**We still use the linear regression framework, but create quadratic, cubic etc. features**

```python
from sklearn.preprocessing import PolynomialFeatures

deg = 3
poly_feats = PolynomialFeatures(degree=deg, include_bias=False)

X_deg = poly_feats.fit_transform(X)

X_poly_feat_names = poly_feats.get_feature_names_out(['x'])

df_X_deg = pd.DataFrame(X_deg, columns=X_poly_feat_names)
df_X_deg.head()
```

```python
X_deg_train, X_deg_test, y_train, y_test = train_test_split(df_X_deg, y, test_size=0.2, random_state=14)

lr_poly = LinearRegression()

lr_poly.fit(X_deg_train, y_train)

lr_poly.coef_
```

```python
lr_poly.intercept_
```

```python
## FROM PREVIOUS LINEAR REGRESSION MODEL
R2_train = lr_model.score(X_train,y_train)
print(f'R2 train for linear: {R2_train}')

R2_test = lr_model.score(X_test,y_test)
print(f'R2 test for linear: {R2_test}')
```

```python
## Polynomial model
R2_train_poly = lr_poly.score(X_deg_train,y_train)
print(f'R2 train for poly: {R2_train_poly}')

R2_test_poly = lr_poly.score(X_deg_test,y_test)
print(f'R2 test for poly: {R2_test_poly}')
```

We see a slight improvement in R2 test for the polynomial model...

## Model Evaluation
Ok, we've trained a model that can make price predictions. Does it give good predictions? Bad? How can we evaluate it?

Are we underfitting? Overfitting?

Remember, we want a model that does well on the **testing** data!



```python
R2_lr_train = lr.score(X_train,y_train)
print(f'R2 lr train: {R2_lr_train}')

R2_lr_test = lr.score(X_test,y_test)
print(f'R2 lr test: {R2_lr_test}')
```

Perhaps underfitting. Try a more complex model?

```python
deg = 2  #increase this for increased complexity
poly_feats = PolynomialFeatures(degree=deg, include_bias=False)

X_deg = poly_feats.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_deg, y, test_size=0.2, random_state=0)

lr_poly = LinearRegression()
lr_poly.fit(X_train, y_train)

R2_lr_poly_train = lr_poly.score(X_train,y_train)
print(f'R2 lr_poly train: {R2_lr_poly_train}')

R2_lr_poly_test = lr_poly.score(X_test,y_test)
print(f'R2 lr_poly test: {R2_lr_poly_test}')
```

Now we are overfitting! Using 44 features instead of the original 8

```python
pd.DataFrame(X_deg, columns=poly_feats.get_feature_names_out()).head()
```

Better performance be achieved by:
- Finding a model somewhere in between (in terms of complexity)

## Bonus: Classification Example

Below is an example of using ML to predict if a tumor is malignant (0) or benign (1), based on tumor features.

[Documentation](https://scikit-learn.org/stable/datasets/toy_dataset.html#breast-cancer-dataset)

```python
from sklearn.datasets import load_breast_cancer

X, y = load_breast_cancer(return_X_y=True, as_frame=True)

X.head()
```

```python
y.value_counts()
```

```python
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

acc_train = log_reg.score(X_train,y_train)
print(f'acc train: {acc_train}')

acc_test = log_reg.score(X_test,y_test)
print(f'acc test: {acc_test}')
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_class = DecisionTreeClassifier(max_depth=10) #change to vary complexity of decision tree
dt_class.fit(X_train, y_train)

acc_train = dt_class.score(X_train,y_train)
print(f'acc train: {acc_train}')

acc_test = dt_class.score(X_test,y_test)
print(f'acc test: {acc_test}')
```
